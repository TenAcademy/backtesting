`Desing a strategy where  we use a Signal value of 1 to mean buy, -1 to mean we sell, and 0 we will  do nothing.
    `

```python
class SignalStrategy(Strategy):
    '''
    use a Signal value of 1 to mean buy, -1 is sell, 
    and 0 is do nothing.
    '''
    def init(self):
        pass

    def next(self):
        current_signal= self.data.Signal[-1]
        if current_signal == 1:
            if not self.position:
                self.buy()
        elif current_signal == -1:
            if self.position:
                self.position.close()
```

```python
bt = Backtest(GOOG, SignalStrategy, cash=10_000)
```

`print the statistics`

```python
stats = bt.run()
print(stats)
```
